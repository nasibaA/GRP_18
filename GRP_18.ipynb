```python
import pandas as pd
import torch
diabetes_df = pd.read_csv('diabetes.csv')
diabetes_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = diabetes_df.drop('Outcome',axis=1).values
y = diabetes_df['Outcome'].values

# Split into training and test set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state=42,stratify=y)

# Standardize
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
print(X_train)
print(X_test)
print(y_train)
```

```python
import torch.nn as nn
import torch.nn.functional as F # where the activation functions are 
# create tensors = matrices 
X_train = torch.FloatTensor(X_train)
print(X_train)
X_test = torch.FloatTensor(X_test)
print(X_test)
```

```python
y_train = torch.LongTensor(y_train)
print(y_train)
y_test = torch.LongTensor(y_test)
print(y_test)
```

```python
#artificial neural network
class ANN_model(nn.Module):
    def __init__(self,input_features=8,hidden1=20,hidden2=20,out_features=2):
        super().__init__()# super is a computed indirect reference. so, it isolates changes
        # and make sure that children in the layers of multiple inheritence are calling
        # the right parents
        self.layer_1_connection = nn.Linear(input_features,hidden1)
        self.layer_2_connection = nn.Linear(hidden1,hidden2)
        self.out = nn.Linear(hidden2,out_features)
     
    def forward(self,x):
        #apply activation functions
        x = F.relu(self.layer_1_connection(x))
        x = F.relu(self.layer_2_connection(x))
        x = self.out(x)
        return x
```

```python
torch.manual_seed(42)
# create instance of model
ann = ANN_model()
```

```python
# loss function
loss_function = nn.CrossEntropyLoss()

#optimizer
optimizer = torch.optim.Adamax(ann.parameters(),lr=0.60) #  learning rate-
                                                    #How fast does it go to learn? There are reasons why you wouldn't want it to learn too quickly 'cause it can make hasty generalizations. You know two slowly is going to take forever, and that's also not good.
```

```python
# run model through multiple epochs/iterations
final_loss = []
n_epochs = 500
for epoch in range(n_epochs):
    y_pred = ann.forward(X_train)
    loss = loss_function(y_pred,y_train)
    final_loss.append(loss)
    
    if epoch % 10 == 1:
        print(f'Epoch number:{epoch} with loss: {loss}')
        
    optimizer.zero_grad()# zero the gradient  before running backwards propagation
    loss.backward()
    optimizer.step() # perform one optimization step each epoch
```

```python
# predictions
y_pred = []

with torch.no_grad():
    for i, data in enumerate(X_test):
        prediction = ann(data)
        y_pred.append(prediction.argmax())
y_pred        
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

I used Adamax optimizer and changed the learning rate from 0.01 to 0.60. I saw on classification report that recall for class 1 and precision for class 0 increased while recall for class 0 and precision for class 1 decreased if I compare to similar results from Adam optimizer. Recall deltas were higher than Precision deltas. 

Write a function that lists and counts the number of divisors for an input value.
Example 1:
Input: 5
Output: “There are 2 divisors: 1 and 5”
Example 2:
Input: 40
Output: “There are 8 divisors: 1, 2, 4, 5, 8, 10, 20, and 40”

```python
def divisor(n):
    """ this function will find divisor numbers """
    list_num = []   
    for i in range(n):
        if n%(i+1)==0:
            list_num.append(i+1)
    return list_num
                  



print(divisor(15))
print(divisor(40))
```
